### The .apply() Method with Row Values

```
bond = pd.read_csv(root + 'jamesbond.csv' , index_col = 'Film')
bond.sort_index(inplace = True)
bond.head()
```

```
bond.loc['A View to a Kill']['Actor']
```

```
def good_movie(row):
    actor = row[1]
    # actor = row['Actor']
    budget = row[4]
    
    if actor == 'Pierce Brodnan':
        return "The Best"
    
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    
    else:
        return "No Idea"
```

```
bond['good movie'] = bond.apply(good_movie , axis = 'columns')
bond.head()
```

```
def good_movie(row , best_actor = 'Pierce Brodnan'):
    actor = row[1]
    # actor = row['Actor']
    budget = row[4]
    
    if actor == best_actor:
        return "The Best"
    
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    
    else:
        return "No Idea"
```

```
bond['good movie_args'] = bond.apply(good_movie , axis = 'columns' , args =('Pierce Brodnan',))
bond.head()
```

```
bond['good movie_args1'] = bond.apply(good_movie , axis = 'columns' , best_actor = 'Pierce Brodnan')
bond.head()
```

```
bond['good movie_args2'] = bond.apply(lambda row: good_movie(row ,best_actor = 'Pierce Brodnan') , axis = 'columns')
bond.head()
```
